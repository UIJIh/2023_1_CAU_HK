```python
 #-*- coding:utf-8 -*-
import urllib3
import json
     
#     // 언어 분석 기술 문어/구어 중 한가지만 선택해 사용
#     // 언어 분석 기술(문어)
openApiURL = "http://aiopen.etri.re.kr:8000/WiseNLU" 


accessKey = "c0d741d0-1225-41da-b291-b1f6039778ae"
analysisCode = "wsd"
text = "회계학과, 재무관리학과, 경영학과, 경제학과, 금융학과, 경영정보학과, 기업회계학과, 세무학과, 재무학과, 화재보험학과"

requestJson = {  
    "argument": {
        "text": text,
        "analysis_code": analysisCode
    }
}

http = urllib3.PoolManager()
response = http.request(
    "POST",
    openApiURL,
    headers={"Content-Type": "application/json; charset=UTF-8", "Authorization" :  accessKey},
    body=json.dumps(requestJson, indent=4)
)

print("[responseCode] " + str(response.status))
print("[responBody]")
print(str(response.data,"utf-8"))
# json.dumps(json.loads(response.data), indent=4, ensure_ascii=False)
```

```python
data = response.data

# JSON 데이터 파싱
parsed_data = json.loads(data)

# "morp" 키에서 "type"이 "NNG"이면서 "lemma"이 "학과"나 "과"가 아닌 행들만 추출
filtered_morp = [item for item in parsed_data['return_object']['sentence'][0]['morp'] if item['type'] == 'NNG' and item['lemma'] not in ['학과', '과']]

# 결과 출력
print(filtered_morp)
```

```python
# "morp" 키에서 "type"이 "NNG"이면서 "lemma"이 "학과"나 "과"가 아닌 행들만 리스트로 만들기
filtered_morp = [item['lemma'] for item in parsed_data['return_object']['sentence'][0]['morp'] if item['type'] == 'NNG' and item['lemma'] not in ['학과', '과']]

# 결과 출력
print(filtered_morp)
```

```python
import pandas as pd
data = pd.read_csv('major_uni.csv')
print(data.head())
```

```python
# 학교 리스트
overlapping_schools = []


# '학부_과(전공)명' 열의 각 값과 filtered_morp 리스트를 비교하여 겹치는 경우 학교명을 리스트에 추가
for major_uni in data['학부_과(전공)명']:
    for word in filtered_morp:
        if word in major_uni:
            overlapping_schools.append(data[data['학부_과(전공)명'] == major_uni]['학교명'].iloc[0])
            break  

unique_overlapping_schools = list(set(overlapping_schools))

print(unique_overlapping_schools)
```

```python
#리스트 문자열 정리(여러개 묶여있는 학교들을 쉼표 기준으로 분리)

universities = []
for item in unique_overlapping_schools:
    universities.extend(item.strip().split(', '))
    
print(universities)
```

```python
import random
# 랜덤하게 5개 대학교 선택
random_schools = random.sample(universities, 5)

# 5개 제외 나머지 대학교 숫자
remaining_count = len(universities) - 5

# 결과 출력
print("랜덤하게 선택된 5개의 대학교:")
print(random_schools)
print("\n나머지 학교의 개수:", remaining_count)
```
